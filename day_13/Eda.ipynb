# EDA = exploratory data analysis
a method use to summarise and analize data_set.in python you can use pandas liberaries to perform eda on data sets.

```python
# import liabraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# import dataset
df = sns.load_dataset('titanic')
```

```python
df.head()
```

```python
df.tail()
```

```python
df.info()
```

```python

# get dtype and shape from info 
df.dtypes
```

```python
# to summarize(summary statistics) i-e all numeric values
df.describe()
```

```python
# find null values i-e where null= true
df.isnull()
```

```python
# summ of all null values
df.isnull().sum()
```

```python
# count of null / total values * 100 = percetage
df.isnull().sum()/ len(df) *100  # following dmas rule
```

```python
# finding unique values i.e number of unique values
df.nunique()
```

```python
df.columns
```

```python
# visualize the null values in heat map
sns.heatmap(df.isnull(), yticklabels=False, cbar=False,cmap='viridis')
```

```python
df.drop(columns='deck', inplace=True)
```

```python
df.columns
```

```python
df.shape
```

```python
# visualize and explore
# analyze = andaza, 
sns.countplot(data=df, x='survived')
```

```python
sns.barplot(data = df, x='survived', y='fare', hue='who')
```

```python
sns.countplot(x='parch', data=df, hue='who')
```

```python
# data distribution [hist plot is used]
sns.histplot(data=df,x='age', bins=50)
```

```python
# alpha = opacity i.e dark/light the color
df['age'].hist(bins=34, color='darkred', alpha=0.5)
```

```python

sns.boxplot(data=df, y='age', x='survived')
```

```python
# jb boxplot banae tu aik side numerical hu and other side categorical
# same as hist ma sirf numerical e show hngy
# count plot ma count khud hoga , categorical value dena ha
sns.boxplot(data=df, x='survived', y='age', hue='who')
```

```python
# scatterplot .. check distribution of data
sns.scatterplot(data=df, x='age', y='fare', hue='pclass')
plt.ylim(0,300)
```

```python
sns.pairplot(df, hue='survived')
```

```python
#stackedbor plot
df.groupby(['pclass','who'])['survived'].mean().unstack().plot(kind='bar', stacked=True)
```

```python
#swarm plot = 
sns.swarmplot(x='survived', y='fare', data=df)
```

```python
sns.swarmplot(x='survived', y='age', data=df, hue='who')
```

```python
sns.catplot(x='survived', y='fare', data=df, hue='embark_town', kind='point')
```

```python
# to check relation of two variables = lmplot
sns.lmplot(data=df, x='age', y='fare', hue='pclass')
#assighnment = set lim insidde y<300..
```

```python
# grouping
#data wrangling starts
# how to add a column and make an age group in dataframe
sns.histplot(data=df, x='age')

```
