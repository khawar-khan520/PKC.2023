```python
import tensorflow as tf
from tensorflow import keras

# Create a sequential model
model = keras.Sequential()

# Add a Dense layer with 1 neuron and 1 input dimension
model.add(keras.layers.Dense(units=5, input_shape=[1]))

# Compile the model with a mean squared error loss function and an Adam optimizer
model.compile(loss='mse', optimizer=tf.keras.optimizers.Adam(learning_rate=0.01))

# Train the model on some input data and labels
model.fit(x=[0, 1, 2, 3, 4], y=[1, 3, 5, 7, 9], epochs=100)
```

```python
print(model.predict([11,20,10]))
```
