```python
# import librairies
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
kashti = sns.load_dataset('titanic')
```

```python
kashti.info()
```

```python
# before plot check null values
kashti.isnull()
```

```python
kashti.isnull().sum()
```

```python
kashti.survived.value_counts().plot(kind="bar")
```

```python
# barplot using sns = seaborn
sns.countplot(x="survived", hue = 'sex',  data = kashti)
```

```python
# scatter plot
# check why hue not working
plt.scatter(x="age", y= "fare", data=kashti, chue= "pclass")
```

```python
sns.scatterplot(x="age", y= "fare", data=kashti, hue= "pclass")
```

```python
sns.scatterplot(x="sex", y= "fare", data=kashti, hue= "pclass")
```

```python
sns.barplot(x="sex", y= "fare", data=kashti, hue= "pclass")
```

```python
sns.pairplot(kashti)
```

```python
kashti.hist(bins= 10,figsize=(9,7),grid=False)
```

```python
sns.boxplot(x ="sex", y='fare', hue ='class', data=kashti)
```

```python
sns.boxenplot(x ="sex", y='fare', hue ='class', data=kashti)
```

```python
# assighnment how to change title in x-axis and y-axis
sns.boxenplot(x ="embark_town", y='age', hue ='class', data=kashti)
```

```python
g = sns.FacetGrid(kashti,hue='survived', col='pclass', margin_titles=True, palette={1:'seagreen',0:'grey'})
g = g.map(plt.scatter,"fare", "age", edgecolor="w").add_legend()
```

```python
# when we change hue it will automatically change canva board
g = sns.FacetGrid(kashti,hue='survived', col='embark_town', margin_titles=True, palette={1:'seagreen',0:'grey'})
g = g.map(plt.scatter,"fare", "age", edgecolor="w").add_legend()
```

```python
sns.stripplot(x ="embark_town", y='age', hue ='class', data=kashti)
```

```python
# heat map
# correlation goes from -0 to +
corr = kashti.corr()
plt.figure(figsize=(10,10))
sns.heatmap(corr)
plt.title("heat of kashti data")
```

```python
sns.violinplot(x ="embark_town", y='age', hue ='class', data=kashti)
```
